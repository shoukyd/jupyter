```python
df_sd = pd.read_excel("test-0.xlsx",parse_dates=['Date'],index_col="Date")
# df_sd.shape
# df_sd.info()
df_sdoriginal=df_sd.copy()
df_sd
```

```python
col_list=df_sdoriginal.columns

for c in col_list:
    df_sd[c+'avg 4'] = (df_sdoriginal[c].rolling(4).mean())
    df_sd[c+'std 4'] = (df_sdoriginal[c].rolling(4).std())
    df_sd[c+'avg 7'] = (df_sdoriginal[c].rolling(7).mean())
    df_sd[c+'std 7'] = (df_sdoriginal[c].rolling(7).std())
```

```python
col_list=df_sdoriginal.columns
for c in col_list:
    a=df_sd[c]
    b=[0]*len(a)
    for i in range(len(a)):
        if a[i] == 0:
            b[i] = b[i-1] + 1
        else:
            b[i] = 0
    #print ('this is the list of consecative 0 in the {} Column {}'.format(c,b) )
    #b=pd.Series(b) 
    df_sd[c.swapcase()]= pd.Series(b, index=df_sd.index)
    df_sd
df_sd[['C-13','c-13']]
```
